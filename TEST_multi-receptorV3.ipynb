```python
import numpy as np
import matplotlib.pyplot as plt
import os

def load_c_star_data(filepath):
    """
    Carica i dati C_star da un file .npz.

    Parametri:
    - filepath: str, percorso al file .npz.

    Ritorna:
    - times: np.ndarray, punti temporali.
    - C_star: np.ndarray, valori di C_star.
    """
    if not os.path.exists(filepath):
        raise FileNotFoundError(f"Il file {filepath} non esiste.")
    
    data = np.load(filepath)
    times = data['times']
    C_star = data['C_star']
    return times, C_star

if __name__ == "__main__":
    # Definisci i percorsi ai dati C_star salvati
    data_files = {
        "\u0394t = +10 ms": "Boh_test/min.npz",
        "\u0394t = -10 ms": "Boh_test/min1.npz"
    }

    plt.figure(figsize=(12, 6))

    # Plot ogni dataset C_star
    for label, filepath in data_files.items():
        try:
            times, C_star = load_c_star_data(filepath)
            plt.plot(times, C_star, label=label)
        except Exception as e:
            print(f"Errore nel caricamento di {filepath}: {e}")

    # Aggiungi linee tratteggiate per theta_d e theta_p
    theta_d = 0.00004  # mol/m^3
    theta_p = 0.000075  # mol/m^3
    theta_d1 = 0.000038
    plt.axhline(y=theta_d, color='red', linestyle='--', label=r'theta_d = 0.000075 mol/m$^3$')
    plt.axhline(y=theta_p, color='blue', linestyle='--', label=r'theta_p = 0.00013 mol/m$^3$')
    plt.axhline(y=theta_d1, color='orange', linestyle='--', label=r'theta_d1 = 0.00013 mol/m$^3$')

    # Personalizza il plot
    plt.title(r'Overlay di $C^*$ per $\Delta t = \pm10\,ms$')
    plt.xlabel('Tempo (ms)')
    plt.ylabel(r'$C^*$')
    plt.legend()
    plt.xlim([0, 20000])  # Imposta i limiti dell'asse x da 0 a 8000 ms
    plt.grid(True)
    plt.tight_layout()
    plt.savefig("Boh_test/C_star_overlay.png")
    plt.show()
```

```python


# Definizione delle variabili di tempo
a = 0
b = 20000  # ms
total_duration = 20000  # ms

# Definizione delle fasi della simulazione
phase1_duration = 5000  # ms
phase2_duration = 10000  # ms
phase3_duration = 5000  # ms

# Definizione della frequenza degli spike
frequency = 10.0  # Hz
interval = 1000.0 / frequency  # ms

# Definizione dell'intervallo di simulazione per ogni step
step_size = 100.0  # ms
num_steps = int(total_duration / step_size)

# Percorso al modulo NESTML

# Impostazione della libreria NEST
nest.ResetKernel()
nest.Install(folder + "target/nestml_module")
nest.set_verbosity("M_ERROR")

# Ottenimento dei tipi di recettori dal modello di neurone
receptors = nest.GetDefaults('iaf_cond_bubu_nestml')['receptor_types']
print("Receptors:", receptors)

# Definizione del base_delay
base_delay = 11.0  # ms (Assicurati che base_delay + delta_t sia sempre positivo)

# Funzione per generare spike times
def generate_spike_times(start_time, duration, frequency):
    interval = 1000.0 / frequency
    spike_times = np.arange(start_time + interval, start_time + duration, interval)
    return spike_times.tolist()

# Funzione per calcolare delta_t in base al tempo corrente
def compute_delta_t(current_time):
    if current_time < phase1_duration:
        return -10  # ms
    elif current_time < phase1_duration + phase2_duration:
        # Gradualmente da -10 ms a +10 ms
        t = current_time - phase1_duration
        return -10 + (20.0 / phase2_duration) * t  # Lineare
    else:
        return +10  # ms

# Funzione per aggiornare i delay delle sinapsi
def update_synapse_delays(connections, new_delay):
    # Per CollocatedSynapses, ogni connessione ha due sinapsi (AMPA e NMDA)
    # Passare una lista con due valori di delay
    delays = [new_delay, new_delay]
    nest.SetStatus(connections, {"delay": delays})

# Creazione dei neuroni post e pre-sinaptici
post_neuron = nest.Create("iaf_cond_bubu_nestml__with_stdp_nestml", 1)
pre_neuron = nest.Create("iaf_cond_bubu_nestml__with_stdp_nestml", 1)

# Creazione del registratore di peso
wr = nest.Create("weight_recorder")
nest.SetDefaults("stdp_nestml__with_iaf_cond_bubu_nestml", {"weight_recorder": wr[0]})

# Definizione delle specifiche sinaptiche iniziali con delta_t = -10 ms
syn_spec_post = nest.CollocatedSynapses(
    {
        'synapse_model': 'stdp_nestml__with_iaf_cond_bubu_nestml',
        'receptor_type': receptors['NMDA'],
        'delay': base_delay - 10  # 11 -10 =1 ms
    },
    {
        'synapse_model': 'stdp_nestml__with_iaf_cond_bubu_nestml',
        'receptor_type': receptors['AMPA'],
        'delay': base_delay - 10  # 1 ms
    }
)

syn_spec_pre = nest.CollocatedSynapses(
    {
        'synapse_model': 'static_synapse',
        'receptor_type': receptors['INH_SPIKES'],
        'weight': 999.0,
        'delay': base_delay - 10  # 1 ms
    },
    {
        'synapse_model': 'static_synapse',
        'receptor_type': receptors['EXC_SPIKES'],
        'weight': 999.0,
        'delay': base_delay - 10  # 1 ms
    }
)

# Connessione iniziale del pre_neuron al post_neuron con syn_spec_post
connections_phase1 = nest.Connect(pre_neuron, post_neuron, "one_to_one", syn_spec=syn_spec_post)

# Ottieni le connessioni come SynapseCollection usando nest.GetConnections
connections_phase1 = nest.GetConnections(pre_neuron, post_neuron)

# Creazione di due spike generators distinti
spike_generator_pre = nest.Create("spike_generator")
spike_generator_post = nest.Create("spike_generator")

# Connessione dei spike generators ai neuroni con specifiche sinaptiche appropriate
connections_spike_pre = nest.Connect(spike_generator_pre, pre_neuron, "one_to_one", syn_spec=syn_spec_pre)
connections_spike_post = nest.Connect(spike_generator_post, post_neuron, "one_to_one", syn_spec=syn_spec_post)

# Ottieni le connessioni come SynapseCollection usando nest.GetConnections
connections_spike_pre = nest.GetConnections(spike_generator_pre, pre_neuron)
connections_spike_post = nest.GetConnections(spike_generator_post, post_neuron)

# Creazione dei registratori e degli strumenti di misura
sd = nest.Create("spike_recorder")
nest.Connect(pre_neuron, sd)
nest.Connect(post_neuron, sd)

vm_pre = nest.Create("voltmeter")
vm_post = nest.Create("voltmeter")
nest.Connect(vm_pre, pre_neuron)
nest.Connect(vm_post, post_neuron)

Multi = nest.Create('multimeter', {
    "record_from": [
        "Ica_VDCC", "Eca_syn", "gca_VDCC", "h_infinity",
        "m_infinity", "C_star", "Ca_concentrazione_i", "Rho",
        "Heaviside_p", "I_syn_exc", "g_AMPA__X__AMPA", "I_spike",
        "Heaviside_d", "g_AMPA_target", "g_NMDA__X__NMDA",
        "g_AMPA_scale", "I_syn_ampa", "ica_NMDA", "I_syn_nmda"
    ]
})
nest.Connect(Multi, post_neuron)

# Directory di output
output_dir = "Boh_test"
os.makedirs(output_dir, exist_ok=True)

# Simulazione passo-passo
current_time = 0.0
for step in range(num_steps):
    print(f"Inizio Step {step+1}/{num_steps}: Tempo {current_time} ms")
    
    # Calcola delta_t per il tempo corrente
    delta_t = compute_delta_t(current_time)
    new_delay = base_delay + delta_t
    
    if new_delay <= 0:
        raise ValueError(f"Il delay non pu essere negativo o zero. Tempo: {current_time} ms, Delay: {new_delay} ms")
    
    # Aggiorna i delay delle connessioni
    update_synapse_delays(connections_phase1, new_delay)
    update_synapse_delays(connections_spike_pre, new_delay)
    update_synapse_delays(connections_spike_post, new_delay)
    
    # Genera spike times per questo step
    spike_times_step_pre = generate_spike_times(current_time, step_size, frequency)
    spike_times_step_post = generate_spike_times(current_time, step_size, frequency)
    
    nest.SetStatus(spike_generator_pre, {"spike_times": spike_times_step_pre})
    nest.SetStatus(spike_generator_post, {"spike_times": spike_times_step_post})
    
    # Simula per lo step_size
    nest.Simulate(step_size)
    
    # Aggiorna il tempo corrente
    current_time += step_size

# Raccolta degli spike registrati
spikes = nest.GetStatus(sd, "events")[0]
spike_times = spikes["times"]
spike_senders = spikes["senders"]

# Raccolta dei dati dai multimetri
Ca_concentrazione_i_times = Multi.events["times"]
Ca_concentrazione_i_values = Multi.events["Ca_concentrazione_i"]

c_star_times = Multi.events["times"]
c_star_values = Multi.events["C_star"]

# Salvataggio dei dati
output_filename1 = os.path.join(output_dir, "min3ca.npz")
output_filename2 = os.path.join(output_dir, "min.npz")

np.savez(output_filename1, times=Ca_concentrazione_i_times, Ca_concentrazione_i=Ca_concentrazione_i_values)
np.savez(output_filename2, times=c_star_times, C_star=c_star_values)
print(f"Dati salvati in {output_filename1} e {output_filename2}")

# Plot dei risultati
fig, ax = plt.subplots(6, figsize=(18, 20))

# Potenziale di membrana
ax[0].plot(vm_pre.events["times"], vm_pre.events["V_m"], color="b", label="Pre-sinaptico")
ax[0].plot(vm_post.events["times"], vm_post.events["V_m"], color="orange", label="Post-sinaptico")
ax[0].set_title("Potenziale del neurone post-sinaptico")
ax[0].set_ylabel("V_m [mV]")
ax[0].set_xlim([a, b])
ax[0].legend()

# Raster plot degli spike
ax[1].set_title("Rasterplot")
ax[1].plot(
    spike_times[np.where(spike_senders == post_neuron[0])],
    spike_senders[np.where(spike_senders == post_neuron[0])],
    '.', markersize=10, alpha=0.5, color="orange", label="Spike Post-sinaptico"
)
ax[1].plot(
    spike_times[np.where(spike_senders == pre_neuron[0])],
    spike_senders[np.where(spike_senders == pre_neuron[0])] - 1,
    '.', markersize=10, alpha=0.5, color="b", label="Spike Pre-sinaptico"
)
ax[1].legend()
ax[1].set_ylabel("Neuroni")
ax[1].set_yticks([])
ax[1].set_xlim([a, b])

# Registrazione dei pesi (STDP)
ax[2].plot(wr.events["times"], wr.events["weights"], color='r')
ax[2].set_title("STDP")
ax[2].set_ylabel("Peso [nS]")
ax[2].set_xlabel("Tempo [ms]")
ax[2].set_xlim([a, b])

# I_syn_ampa
ax[3].plot(Multi.events["times"], Multi.events["I_syn_ampa"], color="b", label="I_syn_ampa")
ax[3].set_title("I_syn_ampa")
ax[3].set_ylabel("I_syn_ampa [nA]")
ax[3].set_xlim([a, b])
ax[3].legend()

# g_AMPA_scale
ax[4].plot(Multi.events["times"], Multi.events["g_AMPA_scale"], color="g", label="g_AMPA_scale")
ax[4].set_title("g_AMPA_scale")
ax[4].set_ylabel("g_AMPA_scale")
ax[4].set_xlim([a, b])
ax[4].legend()

# Eca_syn
ax[5].plot(Multi.events["times"], Multi.events["Eca_syn"], color="m", label="Eca_syn")
ax[5].set_title("Eca_syn")
ax[5].set_ylabel("Eca_syn [mV]")
ax[5].set_xlim([a, b])
ax[5].legend()

fig.tight_layout()
plt.show()
```
