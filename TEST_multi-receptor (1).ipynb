# Now we can use the new module

```python
delta_t = 10  # ms
```

```python
#os.environ['LD_LIBRARY_PATH'] = '/tmp/nestml_target:' + os.environ.get('LD_LIBRARY_PATH', '')
nest.ResetKernel()
nest.Install(folder + "target/nestml_module")
nest.set_verbosity("M_ERROR")

duration = 10000 # ms
frequency = 1.0  # HzNESTCodeGeneratorUtils.generate_code_for()
receptors = nest.GetDefaults('iaf_cond_bubu_nestml')['receptor_types']
print(receptors)
syn_spec = nest.CollocatedSynapses(
                                   {'synapse_model': 'stdp_nestml__with_iaf_cond_bubu_nestml','receptor_type': receptors['AMPA']},
                                   {'synapse_model': 'stdp_nestml__with_iaf_cond_bubu_nestml','receptor_type': receptors['AMPA']},
                                   {'synapse_model': 'stdp_nestml__with_iaf_cond_bubu_nestml','receptor_type': receptors['EXC_SPIKES']},
                                   {'synapse_model': 'stdp_nestml__with_iaf_cond_bubu_nestml','receptor_type': receptors['INH_SPIKES']})

post_neuron = nest.Create("iaf_cond_bubu_nestml__with_stdp_nestml", 1)
pre_neuron = nest.Create("iaf_cond_bubu_nestml__with_stdp_nestml", 1)


wr = nest.Create("weight_recorder")
#nest.SetDefaults("static_synapse", {"weight_recorder": wr[0]})
nest.SetDefaults("stdp_nestml__with_iaf_cond_bubu_nestml", {"weight_recorder": wr[0]})

#nest.Connect(pre_neuron, post_neuron, "one_to_one", "static_synapse")
nest.Connect(pre_neuron, post_neuron, "one_to_one", syn_spec=syn_spec)

interval = 1000.0 / frequency
spike_times = np.arange(interval, duration, interval)
spike_generator = nest.Create("spike_generator", {"spike_times": spike_times})

if delta_t >= 0:
    delta_pre = 0.0
    delta_post = delta_t

else:
    delta_pre = -1 * delta_t
    delta_post = 0.0
    
syn_spec2 = nest.CollocatedSynapses(
                                   {'synapse_model': 'static_synapse','receptor_type': receptors['AMPA'], 'weight': 999.0,
                                                          "delay": 1.0 + delta_pre},
                                   {'synapse_model': 'static_synapse','receptor_type': receptors['NMDA'], 'weight': 999.0,
                                                          "delay": 1.0 + delta_pre},
                                   {'synapse_model': 'static_synapse','receptor_type': receptors['INH_SPIKES'], 'weight': 999.0,
                                                          "delay": 1.0 + delta_pre},
                                   {'synapse_model': 'static_synapse','receptor_type': receptors['EXC_SPIKES'],'weight': 999.0,
                                                          'delay': 1.0 + delta_pre})
syn_spec3 = nest.CollocatedSynapses(
                                   {'synapse_model': 'static_synapse','receptor_type': receptors['AMPA'], 'weight': 999.0,
                                                          "delay": 1.0 + delta_post},
                                   {'synapse_model': 'static_synapse','receptor_type': receptors['INH_SPIKES'], 'weight': 999.0,
                                                          "delay": 1.0 + delta_post},
                                   {'synapse_model': 'static_synapse','receptor_type': receptors['EXC_SPIKES'], 'weight': 999.0,
                                                          "delay": 1.0 + delta_post},
                                   {'synapse_model': 'static_synapse','receptor_type': receptors['NMDA'],'weight': 999.0,
                                                          'delay': 1.0 + delta_post})
nest.Connect(spike_generator, pre_neuron, "one_to_one", syn_spec=syn_spec2)

nest.Connect(spike_generator, post_neuron, "one_to_one", syn_spec=syn_spec3)
# Verifica le variabili registrabili
recordables = nest.GetStatus(post_neuron, 'recordables')
print("Variabili registrabili per post_neuron:", recordables)
sd = nest.Create("spike_recorder")
nest.Connect(pre_neuron, sd)
nest.Connect(post_neuron, sd)

vm_pre = nest.Create("voltmeter")
vm_post = nest.Create("voltmeter")
nest.Connect(vm_pre, pre_neuron)
nest.Connect(vm_post, post_neuron)
Multi = nest.Create('multimeter',{"record_from": [ "Ica_VDCC","gca_VDCC","h_infinity","m_infinity", "C_star", "Ca_concentrazione_i", "Rho", "Heaviside_p","Heaviside_d"]  } )
nest.Connect(Multi, post_neuron)

nest.Simulate(duration)
spikes = sd.get(['events'][0])

#print(sd.events)
fig, ax = plt.subplots(5, figsize = (18,10))

ax[0].plot(vm_pre.events["times"], vm_pre.events["V_m"], color = "b", label = "Pre-synaptic potential")
ax[0].plot(vm_post.events["times"], vm_post.events["V_m"], color = "orange",  label = "Post-synaptic potential")
ax[0].set_title("Post-synptic neuron Potential")
ax[0].set_ylabel("Vm [mV]")
ax[0].set_xlim([1800,3500])
ax[0].legend()

ax[1].set_title("Rasterplot")
ax[1].plot( spikes['times'][np.where(spikes['senders']==1)], spikes['senders'][np.where(spikes['senders']==1)], '.', markersize = 40, alpha = 0.5, color = "b", label = "Pre-synaptic spikes")
ax[1].plot( spikes['times'][np.where(spikes['senders']==2)], spikes['senders'][np.where(spikes['senders']==2)]-1, '.', markersize = 40, alpha = 0.5,color = "orange", label = "Post-synaptic spikes")
ax[1].legend()
ax[1].set_ylabel("Neurons")
ax[1].set_yticklabels("")

ax[2].plot(wr.events["times"], wr.events["weights"], color = 'r')
ax[2].set_title("STDP")
ax[2].set_ylabel("Weight [nS]")
ax[2].set_xlabel("Time [ms]")

ax[3].plot(Multi.events["times"], Multi.events["C_star"], color = "b", label = "Pre-synaptic potential")
ax[3].plot(Multi.events["times"], Multi.events["C_star"], color = "orange",  label = "Post-synaptic potential")
ax[3].set_title("C_star")
ax[3].set_ylabel("C_star [non]")
ax[3].set_xlim([0,10000])
ax[3].legend()

ax[4].plot(Multi.events["times"], Multi.events["Rho"], color = "b", label = "Pre-synaptic potential")
ax[4].plot(Multi.events["times"], Multi.events["Rho"], color = "orange",  label = "Post-synaptic potential")
ax[4].set_title("Rho")
ax[4].set_ylabel("Rho [non]")
ax[4].set_xlim(0,10000)
ax[4].legend()



fig.tight_layout()
```
